```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from pywaffle import Waffle
```

```python
xpoint=np.array([1,2,3,4,5])
ypoints=np.array([4,5,6,7,8])
xpoint1=np.array([10,0,30,40,50])
ypoints1=np.array([8,65,4,75,88])
plt.plot(xpoint,ypoints)
plt.plot(xpoint1,ypoints1)
plt.show()
```

```python
fig,ax=plt.subplots()
data={'year':[21,22,23,24,25,26],'sale':[100,120,150,180,210,240],'expenses':[70,150,120,210,200,270]}
df=pd.DataFrame(data)
ax.stackplot(df['year'],df['sale'],df['expenses'],labels=['s','e'])
ax.legend()
plt.show()
```

```python
fig,ax=plt.subplots()
data={'year':[21,22,23,24,25,26],'sale':[100,120,150,180,210,240],'expenses':[70,150,120,210,200,270]}
df=pd.DataFrame(data)
ax.fill_between(df['year'],df['sale'],df['expenses'])
plt.show()
```

```python
fig,ax=plt.subplots()
data={'year':[21,22,23,24,25,26],'sale':[100,120,150,180,210,240],'expenses':[70,150,120,210,200,270]}
df=pd.DataFrame(data)
ax.fill_between(df['year'],df['sale'],df['expenses'])
ax.plot(df['year'],df['sale'],color='red')
ax.plot(df['year'],df['expenses'],color='green')
plt.show()
```

```python
fig,ax=plt.subplots()
data={'year':[21,22,23,24,25,26],'sale':[100,120,150,180,210,240],'expenses':[70,150,120,210,200,270]}
df=pd.DataFrame(data)
ax.fill_between(df['year'],df['sale'],df['expenses'],where=df.sale>df.expenses,color='black',interpolate=True)
ax.fill_between(df['year'],df['sale'],df['expenses'],where=df.sale<df.expenses,color='orange',interpolate=True)

ax.plot(df['year'],df['sale'],color='red')
ax.plot(df['year'],df['expenses'],color='green')
plt.show()
```
