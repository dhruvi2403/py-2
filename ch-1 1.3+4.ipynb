```python
Q1=df['mpg'].quantile(0.25)
Q3=df['mpg'].quantile(0.75)
IQR=Q3-Q1
UL=Q3+1.5*IQR
LL=Q1-1.5*IQR

#remove the outlier
df=df[(df['mpg']>= LL) & (df['mpg']<= UL)]
```

```python
plt.boxplot(df[['acceleration','mpg','horsepower']])
plt.show()
```

```python
QQ1=df['acceleration'].quantile(0.25)
QQ3=df['acceleration'].quantile(0.75)
IQQR=QQ3-QQ1
ULL=QQ3+1.5*IQQR
LLL=QQ1-1.5*IQQR

#remove the outlier
df=df[(df['acceleration']>= LLL) & (df['acceleration']<= ULL)]
```

```python
plt.boxplot(df[['acceleration','mpg','horsepower']])
plt.show()
```

```python
teams=df3.Team1.unique()
L=[]
for team in teams:
    played=df3[(df3['Team1']==team) | (df3['Team2']==team)].shape[0]
    win=df3[df3.WinningTeam==team].shape[0]
    home_win=df3[(df3.WinningTeam==team)&(df3.Team1==team)].shape[0]
    home_play=df3[df3.Team1==team].shape[0]
    away_win=df3[(df3.WinningTeam==team)&(df3.Team2==team)].shape[0]
    away_play=df3[df3.Team2==team].shape[0]
    L.append([team,played,((win/played)*100),((home_win/home_play)*100),(away_win/away_play)*100])
x=pd.DataFrame(L)
print(x)
```
