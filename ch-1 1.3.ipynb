    Q1 Q2 Q3 (BOTTOM TO TOP)
    Q3-Q1 => IQR

```python
Q1=df['mpg'].quantile(0.25)
Q3=df['mpg'].quantile(0.75)
IQR=Q3-Q1
UL=Q3+1.5*IQR
LL=Q1-1.5*IQR

#remove the outlier
df=df[(df['mpg']>= LL) & (df['mpg']<= UL)]
```

```python
plt.boxplot(df[['acceleration','mpg','horsepower']])
plt.show()
```

```python
QQ1=df['acceleration'].quantile(0.25)
QQ3=df['acceleration'].quantile(0.75)
IQQR=QQ3-QQ1
ULL=QQ3+1.5*IQQR
LLL=QQ1-1.5*IQQR

#remove the outlier
df=df[(df['acceleration']>= LLL) & (df['acceleration']<= ULL)]
```

```python
plt.boxplot(df[['acceleration','mpg','horsepower']])
plt.show()
```
